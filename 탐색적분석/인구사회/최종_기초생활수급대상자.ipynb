# 2018

```python
basic_living_2018 = basic_living['2018_기초수급자현황']
basic_living_2018.head()
```

```python
# 읍면동 명칭 수정

basic_living_2018['emd_nm'] = basic_living_2018['emd_nm'].apply(lambda x :'옥련동' if x in ['옥련동', '옥련1동', '옥련2동'] else x )
basic_living_2018['emd_nm'] = basic_living_2018['emd_nm'].apply(lambda x :'연수동' if x in ['연수동', '연수1동', 
                                                                                         '연수2동', '연수3동'] else x )
basic_living_2018['emd_nm'] = basic_living_2018['emd_nm'].apply(lambda x :'동춘동' if x in ['동춘동', '동춘1동', 
                                                                                         '동춘2동', '동춘3동'] else x )
basic_living_2018['emd_nm'] = basic_living_2018['emd_nm'].apply(lambda x :'송도동' if x in ['송도동','송도1동', '송도2동', 
                                                                           '송도3동', '송도4동', '송도5동'] else x )
```

```python
basic_living_2018 = basic_living_2018[basic_living_2018['emd_nm'] != '합 계'].groupby('emd_nm').sum()
basic_living_2018.reset_index(inplace = True)
basic_living_2018.insert(0, 'base_date','2018')
basic_living_2018
```

# 최종 : 기초생활수급자_신도심_원도심

```python
# 원도심
basic_living_old = basic_living_fin[basic_living_fin['emd_nm'] != '송도동']
basic_living_old = basic_living_old.groupby('base_date').sum()
basic_living_old.reset_index(inplace = True)
basic_living_old.insert(1, 'center', '원도심')
basic_living_old
```

```python
# 원도심
basic_living_new = basic_living_fin[basic_living_fin['emd_nm'] == '송도동']
basic_living_new.reset_index(inplace = True)
basic_living_new = basic_living_new.drop(columns = ['index', 'emd_nm'])
basic_living_new.insert(1, 'center', '신도심')
basic_living_new
```

```python
# 합치기
basic_living_old_new = pd.concat([basic_living_old, basic_living_new])
basic_living_old_new = basic_living_old_new.sort_values(by = basic_living_old_new.columns[0])
basic_living_old_new.reset_index(inplace = True)
basic_living_old_new = basic_living_old_new.drop(columns = 'index')
basic_living_old_new
```

```python
# 저장
basic_living_old_new.to_csv('최종_기초생활수급자_신도심_원도심.csv', encoding = 'utf-8', index = False)
```
