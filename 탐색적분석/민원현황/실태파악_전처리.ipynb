# 민원현황 : Complaint

```python
Complaint = pd.read_excel('9_연수구_민원현황_격자매핑.xlsx')
```

```python
# 결측치 제거
Complaint.dropna(inplace = True)
```

```python
# 인덱스 재설정
Complaint.reset_index(inplace = True)
Complaint.drop('index', axis = 1 , inplace = True)
```

```python
Complaint.head()
```

```python
# 복사
Complaint2 = Complaint.copy()
```

```python
# 민원수 count를 위해 임의로 1의 값만 가지는 칼럼 형성 
Complaint2['count'] = 1
```

```python
# 민원내용은 필요 없으므로 삭제
Complaint2.drop('compl_contents', axis = 1, inplace = True)
```

```python
# 년도만 추출하기 위한 형 변환
Complaint2['apply_date'] = Complaint2['apply_date'].astype('str')

# 년도 추출
temp = []
for i in range(len(Complaint2)):
    temp.append(Complaint2.loc[i]['apply_date'][:4])
    
Complaint2['apply_date'] = temp
```

```python
Complaint2.head()
```

```python
# 년도별 df 생성을 위해 년도별로 리스트에 값 넣어주기
list_21 = []

for i in range(len(Complaint2)):
               
    if Complaint2.loc[i]['apply_date'] == '2021':
        list_21.append(Complaint2.loc[i])
        
# 년도별 딕셔너리 형성
Complaint_year = {'2021' : list_21}
```

```python
# 2021년 df 조작

Complaint_21 = pd.DataFrame(Complaint_year['2021'])
Complaint_21.reset_index(inplace = True)
Complaint_21.drop(['index', 'apply_date'], axis = 1, inplace = True)
Complaint_21 = Complaint_21.groupby('gid').sum()
Complaint_21.rename(columns = {'count' : 'complaint_cnt'}, inplace = True)

Complaint_21.head()
```

```python
Complaint_21.reset_index(inplace = True)
```

```python
Complaint_21.head()
```

```python
# 민원현황 데이터 저장
Complaint_21.to_csv('민원현황_2021.csv', index = False, encoding = 'utf-8')
```

# 고충민원 : Grievance

```python
Grievance = pd.read_excel('8_연수구_고충민원_격자매핑.xlsx')
Grievance.head()
```

```python
# 결측치 제거
Grievance.dropna(inplace = True)
```

```python
# 결측치 제거로 인한 인덱스 재설정

Grievance.reset_index(inplace = True)
Grievance.drop('index', axis = 1 , inplace = True)
```

```python
Grievance2 = Grievance.copy()
```

```python
# 고충민원수 count를 위해 임의로 1의 값만 가지는 칼럼 형성 
Grievance2['count'] = 1
```

```python
# 민원내용은 필요 없으므로 삭제
Grievance2.drop('compl_contents', axis = 1, inplace = True)
```

```python
# 년도만 추출하기 위한 형 변환
Grievance2['apply_date'] = Grievance2['apply_date'].astype('str')

# 년도 추출
temp = []
for i in range(len(Grievance2)):
    temp.append(Grievance2.loc[i]['apply_date'][:4])
    
Grievance2['apply_date'] = temp
```

```python
Grievance2.head()
```

```python
# 년도별 df 생성을 위해 년도별로 리스트에 값 넣어주기
list_21 = []

for i in range(len(Grievance2)):
               
    if Grievance2.loc[i]['apply_date'] == '2021':
        list_21.append(Grievance2.loc[i])
        

# 년도별 딕셔너리 형성
Grievance_year = {'2021' : list_21}
```

```python
grievance_21 = pd.DataFrame(Grievance_year['2021'])
grievance_21.reset_index(inplace = True)
grievance_21.drop(['index', 'apply_date'], axis = 1, inplace = True)
grievance_21.head()
```

```python
# gid 기준으로 민원수 합하기 ! (민원수 합하기)
grievance_21 = Grievance_21.groupby('gid').sum()
grievance_21.head()
```

```python
# 칼럼면 변경 : 'grievance_cnt'
grievance_21.rename(columns = {'count' : 'grievance_cnt'}, inplace = True)
```

```python
grievance_21.reset_index(inplace = True)
```

```python
grievance_21.head()
```

```python
# 고충민원 데이터 저장
grievance_21.to_csv('고충민원_2021.csv', index = False, encoding = 'utf-8')
```

# 거주인구 : population

```python
population_sheet = pd.read_excel('3_연수구_거주인구_격자매핑.xlsx', sheet_name = None )
```

```python
population_sheet.keys()
```

```python
type(population_sheet['2021'])
```

```python
population_21 = population_sheet['2021']
population_21.head()
```

```python
# 2021 거주 인구 총합 'population_cnt'로 저장
population_21['population_cnt'] = population_21.iloc[:][:].sum(axis = 1)
population_21.head()
```

```python
# 필요없는 칼럼삭제
population_21.drop(columns = ['m_0g_pop', 'w_0g_pop','m_20g_pop', 'w_20g_pop', 'm_30g_pop',
                              'w_30g_pop', 'm_40g_pop','w_40g_pop','m_50g_pop','w_50g_pop','m_60g_pop',
                              'w_60g_pop','m_70g_pop','w_70g_pop'], axis = 1, inplace = True)

population_21.head()
```

```python
# 거주인구 데이터 저장
population_21.to_csv('거주인구_2021.csv', index = False, encoding = 'utf-8')
```
