# 2. 데이터 전처리

## 1) 총종사자 증감률

```python
## 데이터프레임의 규칙을 보면, 컬럼이 '단순연도' 인 경우(예.2013, 2014) '총 종사자수'를 의미함. 

# 동이름 컬럼과 함께'총 종사자수' 내용을 담은 새 데이터프레임 생성
employee = df1[['동별(1)','2013','2014','2015','2016','2017','2018']] 

# 불필요한 0행 삭제 
employee.drop(0, inplace=True) 

# 행 삭제에 따른 인덱스 재설정 
employee.reset_index(drop = True, inplace = True) 

# 컬럼명 변경 
employee.rename(columns={'동별(1)':'dong'}, inplace=True) 

employee
```

```python
employee.info()
```

```python
## 문제점

#  2014년 9월 24일에 송도1동에서 송도3동이 분리됨. 
#  2013년 송도3동의 데이터는 모두 null 값

## 그래서 어떻게 할 것인가? 

#  '송도1동 + 송도2동 + 송도3동 = 송도동'으로 계산해서 원도심과 달리 송도는 세 동을 합친 '송도동' 수준에서 보는게 어떨끼?  


# 위와 같은 배경에서, null 값을 0으로 변환
employee.iloc[13,1] = 0

# object -> int64 형변환 
employee = employee.astype({"2013":"int64", "2014":"int64", "2015":"int64", "2016":"int64", "2017":"int64", "2018":"int64"})
employee.info()
employee
```

```python
# '송'으로 시작하는 값을 regex에 할당함 
regex = '송.*'

# 새로운 14행을 생성 = 송도1동 + 송도2동 + 송도3동. 다만, 'dong'열은 제외하고 합산하기. 따라서, 14행 'dong' 열은 null값.
employee.loc[14,1:] = employee[employee.dong.str.match(regex)].sum()

# null 값이 들어 있는 14행 'dong'열에 '송도동' 명칭 부여. 
employee['dong'].replace({None: '송도동'}, inplace=True)
employee
```

```python
okryun = employee.iloc[1:3, 1:].sum()
yeonsu = employee.iloc[4:7, 1:].sum()
dongchun = employee.iloc[8:11, 1:].sum()

employee.loc[15, 1:] = okryun
employee.iloc[15,0] = '옥련동'

employee.loc[16, 1:] = yeonsu
employee.iloc[16,0] = '연수동'

employee.loc[17, 1:] = dongchun
employee.iloc[17,0] = '동춘동'

employee
```

```python
# 필요한 행만 추출하여 데이터프레임 재생성

filter = ['옥련동', '선학동', '연수동', '청학동', '동춘동', '송도동']

employee2 = employee.query("dong == @filter")
employee2.reset_index(drop=True, inplace=True)
employee2
```

```python
# (기준연도 총종사자수 - 비교연도 총종사자수) / 비교연도 총종사자수 * 100
# 기준연도: 2018, 비교연도: 2013
# 산출 근거: 국토교통부(2021, p.12)

employee2['증감률(2013-2018)'] = (employee2['2018'] - employee2['2013'])/employee2['2013']*100
employee2
```

```python
# 데이터 내보내기
employee2.to_csv('최종_산업경제_총종사자증감률_2013-2018.csv')
```

```python
## 문제점

#  2014년 9월 24일에 송도1동에서 송도3동이 분리됨. 
#  2013년 송도3동의 데이터는 모두 null 값

## 그래서 어떻게 할 것인가? 

#  '송도1동 + 송도2동 + 송도3동 = 송도동'으로 계산해서 원도심과 달리 송도는 세 동을 합친 '송도동' 수준에서 보는게 어떨끼?   


# 위와 같은 배경에서, null 값을 0으로 변환
lodging_restaurant.iloc[13,1] = 0

# object -> int64 형변환 
lodging_restaurant =lodging_restaurant.astype({"2013.9":"int64", "2014.9":"int64", "2015.9":"int64", "2016.9":"int64", "2017.9":"int64", "2018.9":"int64"})
lodging_restaurant.info()
lodging_restaurant
```

```python
### 이 코드는 무시하기!!!
# # '송'으로 시작하는 값을 regex에 할당함 
# regex = '송.*'

# # 새로운 14행을 생성 = 송도1동 + 송도2동 + 송도3동. 다만, 'dong'열은 제외하고 합산하기. 따라서, 14행 'dong' 열은 null값.
# lodging_restaurant.loc[14,1:] = lodging_restaurant[lodging_restaurant.dong.str.match(regex)].sum()

# # null 값이 들어 있는 14행 'dong'열에 '송도동' 명칭 부여. 
# lodging_restaurant['dong'].replace({None: '송도동'}, inplace=True)
# lodging_restaurant
```

```python
dongchun = lodging_restaurant.iloc[8:11, 1:].sum()
songdo = lodging_restaurant.iloc[11:, 1:].sum()
yeonsu = lodging_restaurant.iloc[4:7, 1:].sum()
okryun = lodging_restaurant.iloc[1:3, 1:].sum()

lodging_restaurant.loc[14,1:] = dongchun
lodging_restaurant['dong'].replace({None: '동춘동'}, inplace=True)

lodging_restaurant.loc[15,1:] = songdo
lodging_restaurant['dong'].replace({None: '송도동'}, inplace=True)

lodging_restaurant.loc[16,1:] = yeonsu
lodging_restaurant['dong'].replace({None: '연수동'}, inplace=True)

lodging_restaurant.loc[17,1:] = okryun
lodging_restaurant['dong'].replace({None: '옥련동'}, inplace=True)

lodging_restaurant
```

```python
# 컬럼명 정리 

lodging_restaurant.rename(columns={'2013.9':'2013','2014.9':'2014','2015.9':'2015','2016.9':'2016','2017.9':'2017','2018.9':'2018'}, inplace=True)
lodging_restaurant
```

```python
# 필요한 행만 추출하여 데이터프레임 재생성

filter = ['옥련동', '선학동', '연수동', '청학동', '동춘동', '송도동']

lodging_restaurant2 = lodging_restaurant.query("dong == @filter")
lodging_restaurant2.reset_index(drop=True, inplace=True)
lodging_restaurant2
```

```python
# (기준연도 총음식숙박업종사자수 - 비교연도 총음식숙박업종사자수) / 비교연도 총음식숙박업종사자수 * 100
# 기준연도: 2018, 비교연도: 2013
# 산출 근거: 국토교통부(2021, p.12)

lodging_restaurant2['증감률(2013년, 2018년)'] = (lodging_restaurant2['2018'] - lodging_restaurant2['2013'])/lodging_restaurant2['2013']*100
lodging_restaurant2
```
